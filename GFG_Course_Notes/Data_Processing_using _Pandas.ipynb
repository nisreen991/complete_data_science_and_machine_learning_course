## Data Analysis - Automatic Categorical

### 1. Total number of Apps in each Category

```python
categories  = {}
for category in df["Category"].unique():
    count = 0
    for apps in df["Category"]:
        if apps  == category:
            count += 1
            categories[category] = count

print(categories)
```

### 2. Total number of Apps in each Type

```python
types = {}
for name in df["Type"].unique():
    count = 0
    for apps in df["Type"]:
        if apps  == name:
            count += 1
            types[name] = count

print(types)
```

### 3. Total Number of Apps for each Content Rating

```python
types = {}
for name in df["Content Rating"].unique():
    count = 0
    for apps in df["Content Rating"]:
        if apps  == name:
            count += 1
            types[name] = count

print(types)
```

## Removing Null Values from Google Playstore Dataset

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("googleplaystore.csv")
df.head()
df.isnull().sum()
```

```python
# Rating column has large number of null values. using dropna() deletes almost 15% of the data which might give error while analysing data. To overcome this problem, using imputer here.
```

```python
impute = SimpleImputer(missing_values = np.nan, strategy='mean')
x = impute.fit_transform(df.iloc[:,2:3].values)
df.iloc[:,2:3] = x
df.head()
df.isnull().sum()
```

```python
# For remaining null values, dropping them as they are less in number
df = df.dropna()
df.isnull().sum()
df
```

## GroupBy in Pandas

### 1. Name apps that are in ART_AND_DESIGN Category and Rating more than 4.5 in ascending order of their Ratings.

```python
df_pr = df[df['Category'] == "ART_AND_DESIGN"]
df_pr = df_pr[df_pr["Rating"] > 4.5]
df_pr.sort_values(by = "Rating", ascending = True)
```

### 2. Name apps that are in ART_AND_DESIGN Category in descending order of their Reviews.

```python
df_pr = df[df['Category'] == "ART_AND_DESIGN"]
df_pr.sort_values(by = "Reviews", ascending = False)
```

### 3. Which Category is having maximum average Rating?

```python
df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
```

### 4. Count Number of Apps in each Category in descending Order.

```python
df.groupby('Category')['Type'].count().sort_values(ascending=False)
```

### 5. How many Free Apps are there in each Category?

```python
df_pr = df[df['Type'] == 'Free']
df_pr.groupby('Category')['Type'].count().sort_values(ascending=False)
```
