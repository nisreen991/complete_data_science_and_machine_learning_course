### Bill Generation and Updation of Inventory

```python
# Reading from Inventory
fd = open('Inventory.txt', 'r') ##How to read from a file
products = fd.read().split('\n')
fd.close()

# Taking user input
user_prod_id_input = input("Enter Product ID: ")
user_prod_quantity_input = input("Enter Required Quantity: ")
updated_product_list = []

# Going thorugh each product detail
for product in products:
    
    prod_details = product.split(',')

    # Checking if product exists or not
    if prod_details[0] == user_prod_id_input:

        # Checking if we are having enough quantity as requested by user or not
        if int(user_prod_quantity_input) <= int(prod_details[3]):
            print('===================================')
            print('Product Name     :',prod_details[1])
            print('Product Price    :',prod_details[2])
            print('Quantity         :',user_prod_quantity_input)
            print('===================================')
            print('Billing Amount   :',int(user_prod_quantity_input) * int(prod_details[2]))
            print('===================================')
            prod_details[3] = str(int(prod_details[3]) - int(user_prod_quantity_input))

        # If we're not having enough quantity
        else:
            print('Sorry we are not having', user_prod_quantity_input, 'nos. of ', prod_details[1], 'as requested by you.')
            print('We\'re having only ', prod_details[3], ' nos.')
            print('Would you like to purchase it?')
            ch = input('Press Y/N')

            # If user wants to purchase from remaining stock
            if ch == 'Y' or ch =='y':
                print('===================================')
                print('Product Name     :',prod_details[1])
                print('Product Price    :',prod_details[2])
                print('Quantity         :',prod_details[3])
                print('===================================')
                print('Billing Amount   :',int(prod_details[3]) * int(prod_details[2]))
                print('===================================')
                prod_details[3] = '0'
            
    # Updating Inventory List    
    updated_product_list.append(prod_details) #Updating product quantity after purchase

lst = []

# Updating Inventory String
for product in updated_product_list:
    prod = product[0] + "," + product[1] + "," + product[2] + "," + product[3] + '\n'
    lst.append(prod)
lst[-1] = lst[-1][:-1]

# Updating Inventory File
fd = open('Inventory.txt', 'w')

for i in lst:
    fd.write(i)

fd.close()
```

## Generating Sales Transaction Details

```python
import time
# Reading from Inventory
fd = open('Inventory.txt', 'r') ##How to read from a file
products = fd.read().split('\n')
fd.close()

# Taking user input
user_name_input = input('Enter username: ')
user_mail_input = input('Enter user email: ')
user_phone_input = input('Enter user phone number: ')
user_prod_id_input = input("Enter Product ID: ")
user_prod_quantity_input = input("Enter Required Quantity: ")
updated_product_list = []

# Going thorugh each product detail
for product in products:
    
    prod_details = product.split(',')

    # Checking if product exists or not
    if prod_details[0] == user_prod_id_input:

        # Checking if we are having enough quantity as requested by user or not
        if int(user_prod_quantity_input) <= int(prod_details[3]):
            print('===================================')
            print('Product Name     :',prod_details[1])
            print('Product Price    :',prod_details[2])
            print('Quantity         :',user_prod_quantity_input)
            print('===================================')
            print('Billing Amount   :',int(user_prod_quantity_input) * int(prod_details[2]))
            print('===================================')

            # Updating Inventory List
            prod_details[3] = str(int(prod_details[3]) - int(user_prod_quantity_input))

            # Generating Sales details
            fd = open('Sales.txt', 'a')
            sales_detail = user_name_input + "," + user_mail_input + "," + user_phone_input + "," + user_prod_id_input + "," + prod_details[1] + "," + user_prod_quantity_input + "," + str(int(user_prod_quantity_input) * int(prod_details[2])) +","+ time.ctime() + "\n"
            fd.write(sales_detail)
            fd.close()

        # If we're not having enough quantity
        else:
            print('Sorry we are not having', user_prod_quantity_input, 'nos. of ', prod_details[1], 'as requested by you.')
            print('We\'re having only ', prod_details[3], ' nos.')
            print('Would you like to purchase it?')
            ch = input('Press Y/N')

            # If user wants to purchase from remaining stock
            if ch == 'Y' or ch =='y':
                print('===================================')
                print('Product Name     :',prod_details[1])
                print('Product Price    :',prod_details[2])
                print('Quantity         :',prod_details[3])
                print('===================================')
                print('Billing Amount   :',int(prod_details[3]) * int(prod_details[2]))
                print('===================================')

                # Generating Sales details
                fd = open('Sales.txt', 'a')
                sales_detail = user_name_input + "," + user_mail_input + "," + user_phone_input + "," + user_prod_id_input + "," + prod_details[1] + "," + prod_details[3] + "," + str(int(prod_details[3]) * int(prod_details[2])) +","+ time.ctime() + "\n"
                fd.write(sales_detail)
                fd.close()

                # Updating Inventory List
                prod_details[3] = '0'

            else: 
                print('Thanks! We\'ll get back to you when ', prod_details[1], 'is in sufficient stock')
            
    # Updating Inventory List    
    updated_product_list.append(prod_details) #Updating product quantity after purchase

lst = []

# Updating Inventory String
for product in updated_product_list:
    prod = product[0] + "," + product[1] + "," + product[2] + "," + product[3] + '\n'
    lst.append(prod)
lst[-1] = lst[-1][:-1]

# Updating Inventory File
fd = open('Inventory.txt', 'w')

for i in lst:
    fd.write(i)

fd.close()
print('Inventory Updated!')
```
