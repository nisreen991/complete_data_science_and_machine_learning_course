```python
# Loading Inventory List in json file
import json 

fd = open('Records.json', 'w')
record = { 1001: {"Name": "5 Star", "Price": 10, "Quantity":100},
           1002: {"Name": "Milky Bar", "Price": 15, "Quantity":100},
           1003: {"Name": "Snickers", "Price": 25, "Quantity":100},
           1004: {"Name": "Blueberry Cheesecake", "Price": 40, "Quantity":100},
           1005: {"Name": "Chocolate Sundae", "Price": 50, "Quantity":100}
}
js = json.dumps(record)
fd.write(js)
fd.close()
```

```python
import json 
import time

fd = open('Records.json', 'r')
js = fd.read()
fd.close()

record = json.loads(js)

print('=================MENU===============\n')

for key in record.keys():
    print(key, ":", record[key]['Name'], "|", record[key]['Price'], "|", record[key]['Quantity'], "|")

print('=====================================')

user_name_input = input('Enter username: ')
user_mail_input = input('Enter user email: ')
user_phone_input = input('Enter user phone number: ')
user_product_id_input = input('Enter Product ID: ')
user_quantity_input = int(input('Enter Quantity: '))

if record[user_product_id_input]['Quantity'] >= user_quantity_input: 

    record[user_product_id_input]['Quantity'] = record[user_product_id_input]['Quantity'] - user_quantity_input
    
    print('=====================================')
    print('              Bill            \n')
    print("Product Name          : ", record[user_product_id_input]['Name'])
    print("Product Price         : ", record[user_product_id_input]['Price'], "Rs")
    print("Quantity              : ", user_quantity_input)
    
    print('=====================================')
    print("Billing Amount        :", record[user_product_id_input]['Price'] * user_quantity_input, "Rs/-")
    print('=====================================')

    # Generating Sales details
    fd = open('Sales.csv', 'a')
    sales_detail = user_name_input + "," + user_mail_input + "," + user_phone_input + "," + user_product_id_input + "," + record[user_product_id_input]['Name'] + "," + str(user_quantity_input) + "," + str(record[user_product_id_input]['Price'] * user_quantity_input) +","+ time.ctime() + "\n"
    fd.write(sales_detail)
    fd.close()

    print('Thank you for your order! Inventory Updated.')

else:
    print('=====================================')
    print('Sorry! We\'re not having', user_quantity_input, 'nos. of', record[user_product_id_input]['Name'], 'to serve you at the moment.\n')
    print('We\'re left with', record[user_product_id_input]['Quantity'], 'nos. of', record[user_product_id_input]['Name'], '.\n')

    if record[user_product_id_input]['Quantity'] > 0:        
        print('Would you like to purchase it?')
        user_input = input('Press Y/N')

        if user_input == 'Y' or user_input == 'y':
                
            print('=====================================')
            print('              Bill            \n')
            print("Product Name          : ", record[user_product_id_input]['Name'])
            print("Product Price         : ", record[user_product_id_input]['Price'], "Rs")
            print("Quantity              : ", record[user_product_id_input]['Quantity'])
            
            print('=====================================')
            print("Billing Amount        :", record[user_product_id_input]['Price'] * record[user_product_id_input]['Quantity'], "Rs/-")
            print('=====================================')

            # Generating Sales details
            fd = open('Sales.csv', 'a')
            sales_detail = user_name_input + "," + user_mail_input + "," + user_phone_input + "," + user_product_id_input + "," + record[user_product_id_input]['Name'] + "," + str(record[user_product_id_input]['Quantity']) + "," + str(record[user_product_id_input]['Price'] * record[user_product_id_input]['Quantity']) +","+ time.ctime() + "\n"
            fd.write(sales_detail)
            fd.close()
    
            record[user_product_id_input]['Quantity'] = 0
    
            print('=====================================')
            print('Thank you for your order! Inventory Updated.')
    
        else: 
            print('Sorry to let you go!')

    else:
        print('Try again in sometime!')

js = json.dumps(record)
fd = open('Records.json', 'w')
fd.write(js)
fd.close()

print('======================================')
```
