```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    "User-Agent": "Mozilla/5.0"
}
res = requests.get("https://en.wikipedia.org/wiki/Martin_Luther_King_Jr.", headers=headers)
soup = BeautifulSoup(res.text,'html.parser')
```

```python
heading = soup.find('h1').text
```

```python
corpus = ''

for p in soup.find_all('p'):
    corpus += p.text
    corpus += '\n'
    
corpus = corpus.strip()
```

```python
for i in range(3,467):
    corpus = corpus.replace('['+str(i)+']','')
```

```python
fd = open(heading + '.txt', 'w', encoding='utf-8')
fd.write(corpus)
fd.close()
```
