```python
import requests
from bs4 import BeautifulSoup
```

```python
link = 'http://quotes.toscrape.com/'
```

```python
res = requests.get(link)
```

```python
html = res.text
```

```python
fd = open('main.html', 'w', encoding='utf-8')
fd.write(html)
fd.close()
```

# Scraping Quotes

```python
soup = BeautifulSoup(res.text, 'html.parser')
soup
```

```python
soup.find('span', class_ = 'text')
```

```python
soup.find('span', class_ = 'text').text
```

```python
print(soup.find('span', class_ = 'text').text[1:-1]) #removed quotes
```

```python
soup.find_all('span', class_='text')
```

```python
quotes = []
for quote in soup.find_all('span', class_='text'):
    quotes.append(quote.text[1:-1])
    print(quotes)
```

# Scraping Quotes with Author Details

```python
soup.find_all('div', class_ = 'quote')
```

```python
data = []
for sp in soup.find_all('div', class_ = 'quote'):
    quote = sp.find('span',class_ = 'text').text[1:-1]
    author = sp.find('small').text
    author_id = sp.find('a').get('href')

    tags = []
    for tag in sp.find_all('a', class_ = 'tag'):
        tags.append(tag.text)
    tags = ','.join(tags)

    data.append([quote,author,author_id,tags])
print(data)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data, columns = ['quote','author','author_id','tags'])
df
```

```python
df.to_csv('Quotes.csv', index=False)
```

# Scraping Quotes from Multiple Pages

```python
from tqdm import tqdm
```

```python
data = []
for page in tqdm(range(1,11)):
    link = link = 'http://quotes.toscrape.com/page/' + str(page)
    res = requests.get(link)
    soup = BeautifulSoup(res.text, 'html.parser')

    for sp in soup.find_all('div', class_ = 'quote'):
        quote = sp.find('span',class_ = 'text').text[1:-1]
        author = sp.find('small').text
        author_id = sp.find('a').get('href')
    
        tags = []
        for tag in sp.find_all('a', class_ = 'tag'):
            tags.append(tag.text)
        tags = ','.join(tags)
    
        data.append([quote,author,author_id,tags])
    
```

```python
df = pd.DataFrame(data, columns = ['quote','author','author_id','tags'])
df
```

```python
df['author_link'] = 'https://quotes.toscrape.com' + df['author_id']
```

```python
df.to_csv('Quotes.csv', index=False)
```
