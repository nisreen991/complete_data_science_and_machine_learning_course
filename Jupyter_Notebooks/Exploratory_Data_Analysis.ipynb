## 1. Different Methods of Dealing with Missing Data

1. Deleting the column with missing data
2. Deleting the row with missing data
3. Filling the missing values - Imputation
   1. Numerical data - use Mean
   2. Categorical Data
       - use Mode
       - assign NaN values their own Category
4. Advanced Imputation

```python
import pandas as pd
```

```python
df = pd.read_csv("AB_NYC_2019.csv")
```

### 1. Deleting the column with missing data

```python
df1 = df.copy()
```

```python
df1.head()
```

```python
df1.isna().sum()
```

```python
df1.drop("last_review", axis = 1, inplace = True)
```

```python
df1.isna().sum()
```

### 2. Deleting row with missing data

```python
df2 = df.copy()
```

```python
df2.isna().sum()
```

```python
df2.dropna(inplace=True)
```

```python
df2.isna().sum()
```

```python
df2.shape
```

### 3. Filling missing Values using Imputation

#### A) Dealing with Numerical values

```python
df3 = df.copy()
```

```python
df3.head()
```

```python
df3.isna().sum()
```

```python
mean_value = df3["reviews_per_month"].mean()
df3["reviews_per_month"].fillna(mean_value, inplace = True)
```

```python
df3.isna().sum()
```

```python
df3.head()
```

#### B) Dealing with Categorical data

#### i) Use Mode Value

```python
df3.nunique()
```

```python
df3["last_review"].fillna(df3["last_review"].value_counts().index[0], inplace=True)
```

```python
df3.isna().sum()
```

#### ii) Assign a new Category

```python
df4 = df.copy()
```

```python
df4.isna().sum()
```

```python
df4["last_review"].fillna("Not_Reviewed", inplace=True)
```

```python
df4.sample(5)
```

### 4. Advanced Imputation

```python
df5 = df.copy()
```

```python
df5.isna().sum()
```

```python
df5["reviews_per_month"].interpolate(inplace = True)
```

```python
df5.head(5)
```
