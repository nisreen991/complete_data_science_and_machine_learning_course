# Univariate Analysis

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("AB_NYC_2019.csv")
df.head()
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
```

```python
df["id"] = df["id"].astype(str)
df["host_id"] = df["host_id"].astype(str)
df["latitude"] = df["latitude"].astype(str)
df["longitude"] = df["longitude"].astype(str)
```

```python
df.describe()
```

```python
df.nunique()
```

# Bivariate Analysis

## Numerical

```python
df2 = sns.load_dataset("titanic")
df2.head()
```

```python
sns.barplot(x="class", y="age", data=df2, palette="Set2")
```

```python
sns.barplot(x="class", y="fare", data=df2, palette="Set2", estimator="sum", hue="sex")
```

```python
sns.barplot(x="class", y="age", hue="sex", data=df2, palette="deep")
```

```python
sns.boxplot(x="class", y="age", data=df2, palette="pastel")
```

```python
sns.barplot(x="class", y="age", data=df2, palette="Set2", hue="survived")
```

```python
sns.boxplot(x="class", y="age", data=df2, palette="Set2", hue="survived")
```

```python
sns.violinplot(x="class", y="age", data=df2, palette="deep")
```

```python
sns.distplot(df2[df2["class"]=="Second"]["age"])
```

## Categorical

```python
sns.countplot(x="survived", hue="class", data=df2)
```

```python
x = pd.crosstab(df2["survived"],df2["class"])
x
```

```python
sns.heatmap(x, annot=True, fmt=".1f", cmap="coolwarm")
```

```python
sns.stripplot(x = "class", y="age", hue="sex", data=df2)
```
