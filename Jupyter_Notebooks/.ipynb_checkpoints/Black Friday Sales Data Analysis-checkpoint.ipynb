# Analysing Gender Column

## Number of males and females

```python
len(df[df['Gender'] == 'M']) , len(df[df['Gender'] == 'F'])
```

```python
import matplotlib.pyplot as plt
```

```python
# First Approach
data = pd.DataFrame({"Ratio" : [len(df[df['Gender'] == 'M']) , len(df[df['Gender'] == 'F'])]}, index=["Male", "Female"])
data.plot(kind = "pie", autopct="%.2f", y="Ratio")
plt.show()
```

```python
# Second Approach
df.groupby('Gender').size().plot(kind='pie', autopct='%.2f', title='Gender Ratio')
```

```python
df.groupby('Gender').size().plot(kind='bar')
```

## Purchase percentage of each Gender

```python
df.groupby('Gender')['Purchase'].sum().plot(kind="pie", autopct="%.2f")
```

## Per Product Purchase by Each Gender

```python
df.groupby('Gender')['Purchase'].mean().plot(kind="pie", autopct="%.2f")
```

# Analysis of Age and Marital Status

## How much Purchase is made by different Age Groups? 

```python
df.groupby("Age")["Purchase"].size().plot(kind="bar", title="Purchase Distribution by Age")
```

## Amount Spent by each Age Group

```python
df.groupby("Age")['Purchase'].sum().plot(kind="bar", title="Amount Spent by Age")
```

## How much money each Age group spends on Average?

```python
df.groupby("Age")['Purchase'].mean().plot(kind="bar", title="Amount Spent by Age on Average")
```

```python
df.groupby("Age")['Purchase'].mean().plot(kind="pie", autopct="%.2f", title="Amount Spent by Age on Average")
```

## How many Married or unmarried people are making purchases?

```python
df.groupby("Marital_Status")["Purchase"].size().plot(kind="pie", autopct="%.2f")
```

# Combining Gender and Marital Status

```python
l = []
for i in range(len(df)):
    l.append(df['Gender'][i] + "_" + str(df["Marital_Status"][i]))
df['MaritalGender'] = l
```

```python
df.head()
```

```python
sns.countplot(x=df['MaritalGender'], palette = "deep")
```

```python
sns.countplot(x = df['Age'] ,hue=df['MaritalGender'], palette = "deep")
```

```python
sns.countplot(x = df['Product_Category_1'] ,hue=df['MaritalGender'], palette = "deep")
```

```python
sns.countplot(x = df['Stay_In_Current_City_Years'], hue = df['MaritalGender'])
```

```python
sns.countplot(x = df['City_Category'], hue = df['MaritalGender'])
```
